2.  Explain the difference between mutable and immutable data types with examples.

Ans - Mutable means we can change the data after it created. and also say modified the data without creating object.

muttable data type - dictionary, set and list.

=> Immutable means can't change the data after it created. and if we can change the data create a new object.

immutable data type - string, float, integer and tuple


```python
## Example -> Mutable data type

num = [1, 2, 3]

num[0] = 10 # changing the first element

print(num)

```

```python
## Example -> immutable data type

num = (1, 2,3)

num[0] = 10 #we can't change the element

print(num)
```
